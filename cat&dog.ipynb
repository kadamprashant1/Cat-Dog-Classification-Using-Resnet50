```python
import numpy as np
import os
import PIL
from PIL import Image

image_dir="E:/temp/cat&dog/training_set/"

train_cat_dir=os.listdir('E:/temp/cat&dog/training_set/cats/')
train_dog_dir=os.listdir('E:/temp/cat&dog/training_set/dogs/')
print(train_dog_dir[:10])
```

```python
import tensorflow 
from tensorflow import keras
```

```python
from keras.applications import ResNet50
from keras import models
from keras import layers
from keras.optimizers import Adam

```

```python
import cv2
from PIL import Image
dataset=[]
lable=[]
INPUT_SIZE=64

for i, image_name in enumerate(train_cat_dir):
   
    if(image_name.split('.')[1]=='jpg'):
        image = cv2.imread(image_dir+ 'cats/' + image_name)
        image= Image.fromarray(image,'RGB')
        image=image.resize((INPUT_SIZE,INPUT_SIZE))
        dataset.append(np.array(image))
        lable.append(0)

for i, image_nam in enumerate(train_dog_dir):
    if(image_nam.split('.')[1]=='jpg'):
        image = cv2.imread(image_dir+ 'dogs/' + image_nam)
        image= Image.fromarray(image,'RGB')
        image=image.resize((INPUT_SIZE,INPUT_SIZE))
        dataset.append(np.array(image))
        lable.append(0)

dataset=np.array(dataset)
lable=np.array(lable)
```

```python
from locale import normalize
import tensorflow
import numpy as np
import sys
from tensorflow import keras
from keras.utils import to_categorical
from unicodedata import category
from sklearn.model_selection import train_test_split


x_train, x_test, y_train, y_test= train_test_split(dataset, lable, test_size=0.2, random_state=0)



x_train=tensorflow.keras.utils.normalize(x_train, axis=1)
x_test=tensorflow.keras.utils.normalize(x_test, axis=1)


y_train=to_categorical(y_train , num_classes=2)
y_test=to_categorical(y_test , num_classes=2)
```

```python
base_model=tensorflow.keras.applications.ResNet50(
    include_top=False,
    weights="imagenet",
    input_shape=(64,64,3)  
)
```

```python
base_model.summary()
```

```python
from argparse import Action
model =models.Sequential()
model.add(base_model)
model.add(keras.layers.AveragePooling2D())
model.add(keras.layers.BatchNormalization())
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(512,activation='relu'))
model.add(keras.layers.Dense(2,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'])

model.fit(x_train, y_train, batch_size=16, verbose=1, epochs=10, validation_data=(x_test,y_test), shuffle=False)

model.save('cat&dogsepoch10.h5')
```
